```python
import pint
from scipy import constants
import matplotlib.pyplot as plt
import numpy as np
from open_atmos_jupyter_utils import show_plot
```

```python
si = pint.UnitRegistry()
#2*si.kg + 3*si.mg
R_str = constants.R * si.J / si.K / si.mol 
Md = 28.966 * si.kg / si.kmol
R = R_str / Md
```

```python
def rho(p, T):
    return p/R/T

#print(rho(1000*si.hPa, 300*si.k).to_base_units())
print(rho(1000*si.hPa, 300*si.K).to(si.g/(si.cm**3)))
si.setup_matplotlib()
T = np.linspace(280, 320) * si.K
plt.plot(T, rho(1000 * si.hPa, T).to_base_units(), label="gęstość powietrza")
plt.legend()
plt.grid()
plt.gca().xaxis.set_units(si.celsius)
show_plot()
```
